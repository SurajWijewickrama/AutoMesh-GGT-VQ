```python
from transformers import BertTokenizer, BertModel

# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertModel.from_pretrained("bert-base-uncased")
```

```python
import os
import json
import torch
from torch_geometric.data import Data

def preprocess_object(obj):
    # Extract graph data
    vertices = torch.tensor(obj['v'], dtype=torch.float)
    edges = torch.tensor(obj['e'], dtype=torch.long).t()  # Transpose for PyTorch Geometric

    # Encode text label using BERT
    text_label_str = obj['n']
    inputs = tokenizer(text_label_str, return_tensors="pt", padding=True, truncation=True, max_length=32)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    text_label_embedding = outputs.last_hidden_state.mean(dim=1).squeeze(0)  # Pool the embeddings

    node_features = vertices  # Use vertices directly as node features

    # Create graph data
    graph_data = Data(x=node_features, edge_index=edges, text_label=text_label_embedding)
    return graph_data

def load_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

        return preprocess_object(data)
    
def create_dataset(folder_path):
    dataset = []
    file_list = os.listdir(folder_path)
    print(len(file_list))
    for file_name in file_list:
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            data = load_json(file_path)
            dataset.append(data)
            print(file_name)
    return dataset
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, GATConv

# Transformer Encoder
class TransformerTextEncoder(nn.Module):
    def __init__(self, input_dim, embed_dim, num_heads, num_layers):
        super(TransformerTextEncoder, self).__init__()
        self.embedding = nn.Embedding(input_dim, embed_dim)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads),
            num_layers=num_layers
        )

    def forward(self, text_input):
        embedded_text = self.embedding(text_input)
        encoded_text = self.transformer(embedded_text)
        return encoded_text.mean(dim=0)  # Pooling to get latent feature c_text

# Graph Encoder (GNN)
class GraphEncoder(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_layers):
        super(GraphEncoder, self).__init__()
        self.layers = nn.ModuleList()
        self.layers.append(GCNConv(in_channels, hidden_channels))
        for _ in range(num_layers - 2):
            self.layers.append(GCNConv(hidden_channels, hidden_channels))
        self.layers.append(GCNConv(hidden_channels, out_channels))

    def forward(self, x, edge_index):
        for layer in self.layers:
            x = F.relu(layer(x, edge_index))
        return x.mean(dim=0)  # Pooling to get latent feature z_G

# Feature Fusion
class FeatureFusion(nn.Module):
    def __init__(self, text_dim, graph_dim, fused_dim):
        super(FeatureFusion, self).__init__()
        self.fc = nn.Linear(text_dim + graph_dim, fused_dim)

    def forward(self, c_text, z_G):
        print(f"Shape of c_text: {c_text.shape}")
        print(f"Shape of z_G: {z_G.shape}")

        if c_text.ndim == 2 and z_G.ndim == 1:
            z_G = z_G.unsqueeze(0).expand(c_text.size(0), -1)
        
        combined_features = torch.cat([c_text, z_G], dim=-1)
        fused_features = self.fc(combined_features)
        return fused_features

# Vector Quantization (VQ)
class VectorQuantization(nn.Module):
    def __init__(self, codebook_size, embedding_dim):
        super(VectorQuantization, self).__init__()
        self.codebook = nn.Embedding(codebook_size, embedding_dim)
        self.codebook.weight.data.uniform_(-1 / codebook_size, 1 / codebook_size)

    def forward(self, z):
        z_flattened = z.view(-1, z.size(-1))
        distances = (torch.sum(z_flattened**2, dim=1, keepdim=True) 
                     - 2 * torch.matmul(z_flattened, self.codebook.weight.T)
                     + torch.sum(self.codebook.weight**2, dim=1))
        encoding_indices = torch.argmin(distances, dim=1)
        z_q = self.codebook(encoding_indices).view(z.shape)
        return z_q, encoding_indices

# Graph Decoder
class GraphDecoder(nn.Module):
    def __init__(self, latent_dim, node_dim, adj_dim):
        super(GraphDecoder, self).__init__()
        self.node_decoder = nn.Linear(latent_dim, node_dim)
        self.adj_decoder = nn.Linear(latent_dim, adj_dim)

    def forward(self, z):
        reconstructed_nodes = self.node_decoder(z)
        reconstructed_adj = self.adj_decoder(z)
        return reconstructed_nodes, reconstructed_adj

# Transformer Decoder
class TransformerGraphDecoder(nn.Module):
    def __init__(self, embed_dim, num_heads, num_layers, vocab_size):
        super(TransformerGraphDecoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.transformer = nn.TransformerDecoder(
            nn.TransformerDecoderLayer(d_model=embed_dim, nhead=num_heads),
            num_layers=num_layers
        )
        self.fc_out = nn.Linear(embed_dim, vocab_size)

    def forward(self, tgt, memory):
        tgt_embedded = self.embedding(tgt)
        decoded = self.transformer(tgt_embedded, memory)
        output = self.fc_out(decoded)
        return output

# Example Pipeline Integration
def main():
    text_encoder = TransformerTextEncoder(input_dim=1000, embed_dim=128, num_heads=4, num_layers=2)
    graph_encoder = GraphEncoder(in_channels=16, hidden_channels=32, out_channels=64, num_layers=3)
    feature_fusion = FeatureFusion(text_dim=128, graph_dim=64, fused_dim=128)
    vector_quantizer = VectorQuantization(codebook_size=512, embedding_dim=128)
    graph_decoder = GraphDecoder(latent_dim=128, node_dim=16, adj_dim=16)
    graph_token_decoder = TransformerGraphDecoder(embed_dim=128, num_heads=4, num_layers=2, vocab_size=100)

    # Dummy inputs
    text_input = torch.randint(0, 1000, (10, 20))  # (batch, sequence_length)
    graph_x = torch.rand(100, 16)  # (num_nodes, node_features)
    edge_index = torch.randint(0, 100, (2, 200))  # (2, num_edges)

    c_text = text_encoder(text_input)
    z_G = graph_encoder(graph_x, edge_index)
    fused_features = feature_fusion(c_text, z_G)
    z_qG, _ = vector_quantizer(fused_features)
    
    reconstructed_nodes, reconstructed_adj = graph_decoder(z_qG)

    # Generate graph tokens
    # Target tensor
    # Target tensor
    tgt = torch.randint(0, 100, (10, 15))  # [batch_size, sequence_length]

# Truncate z_qG to match tgt's sequence length
    z_qG = z_qG.unsqueeze(0)  # [1, 20, 128]
    z_qG = z_qG[:, :tgt.size(1), :]  # Truncate to [1, 15, 128]
    z_qG = z_qG.expand(tgt.size(0), -1, -1)  # Expand to [10, 15, 128]

    print(f"Shape of tgt: {tgt.shape}")  # [10, 15]
    print(f"Shape of z_qG: {z_qG.shape}")  # [10, 15, 128]

# Pass to graph_token_decoder
    graph_tokens = graph_token_decoder(tgt, z_qG)
    print("Graph Tokens:", graph_tokens)



if __name__ == "__main__":
    main()
```

```python
import torch
from torch_geometric.loader import DataLoader
import torch.nn.functional as F
from torch.optim import Adam

# Assume all your defined classes (Transformers, GNNs, etc.) are imported here.

# Dataset and DataLoader
def load_data_and_split(folder_path):
    dataset = create_dataset(folder_path)
    train_size = int(0.8 * len(dataset))
    val_size = int(0.1 * len(dataset))
    test_size = len(dataset) - train_size - val_size
    train_dataset, val_dataset, test_dataset = torch.utils.data.random_split(
        dataset, [train_size, val_size, test_size]
    )
    return train_dataset, val_dataset, test_dataset

def set_mode(model, mode="train"):
    """
    Set all components of the model to train or eval mode.
    Args:
        model (dict): Dictionary of model components.
        mode (str): "train" or "eval".
    """
    for component in model.values():
        if mode == "train":
            component.train()
        elif mode == "eval":
            component.eval()

# Training Function
def train_one_epoch(model, data_loader, optimizer, device):
    set_mode(model, mode="train")
    total_loss = 0

    for data in data_loader:
        data = data.to(device)

        # Inputs
        text_input = data.text_label  # Use the text_label from the data
        graph_x = data.x  # Node features
        edge_index = data.edge_index  # Edge indices

        # Forward pass
        c_text = model['text_encoder'](text_input)
        z_G = model['graph_encoder'](graph_x, edge_index)
        fused_features = model['feature_fusion'](c_text, z_G)
        z_qG, _ = model['vector_quantizer'](fused_features)
        reconstructed_nodes, reconstructed_adj = model['graph_decoder'](z_qG)

        # Loss: Reconstruction and Token Prediction
        loss_nodes = F.mse_loss(reconstructed_nodes, graph_x)
        loss_adj = F.mse_loss(reconstructed_adj, edge_index.float())  # Approximation
        loss = loss_nodes + loss_adj

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(data_loader)

def evaluate(model, data_loader, device):
    set_mode(model, mode="eval")
    total_loss = 0

    with torch.no_grad():
        for data in data_loader:
            data = data.to(device)

            # Inputs
            text_input = data.text_label  # Use the text_label from the data
            graph_x = data.x  # Node features
            edge_index = data.edge_index  # Edge indices

            # Forward pass
            c_text = model['text_encoder'](text_input)
            z_G = model['graph_encoder'](graph_x, edge_index)
            fused_features = model['feature_fusion'](c_text, z_G)
            z_qG, _ = model['vector_quantizer'](fused_features)
            reconstructed_nodes, reconstructed_adj = model['graph_decoder'](z_qG)

            # Loss: Reconstruction and Token Prediction
            loss_nodes = F.mse_loss(reconstructed_nodes, graph_x)
            loss_adj = F.mse_loss(reconstructed_adj, edge_index.float())  # Approximation
            loss = loss_nodes + loss_adj

            total_loss += loss.item()

    return total_loss / len(data_loader)



# Main Training Loop
def main_training(folder_path, num_epochs=10, batch_size=4, learning_rate=1e-3):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Model components
    model = {
        "text_encoder": TransformerTextEncoder(input_dim=1000, embed_dim=128, num_heads=4, num_layers=2).to(device),
        "graph_encoder": GraphEncoder(in_channels=3, hidden_channels=32, out_channels=64, num_layers=3).to(device),
        "feature_fusion": FeatureFusion(text_dim=128, graph_dim=64, fused_dim=128).to(device),
        "vector_quantizer": VectorQuantization(codebook_size=512, embedding_dim=128).to(device),
        "graph_decoder": GraphDecoder(latent_dim=128, node_dim=3, adj_dim=16).to(device),
    }

    # Optimizer
    optimizer = Adam(
        [param for part in model.values() for param in part.parameters()],
        lr=learning_rate,
    )

    # Load Data
    train_dataset, val_dataset, test_dataset = load_data_and_split(folder_path)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    # Training Loop
    for epoch in range(num_epochs):
        train_loss = train_one_epoch(model, train_loader, optimizer, device)
        val_loss = evaluate(model, val_loader, device)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    print("Training Complete!")

    # Testing
    test_loss = evaluate(model, test_loader, device)
    print(f"Test Loss: {test_loss:.4f}")

# Run the training
folder_path = "./json"  # Update this path
main_training(folder_path)
```
